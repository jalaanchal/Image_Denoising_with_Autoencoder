## Step 3 : Perform Data Preprocessing

```python
#Normalization
X_train = X_train/255
X_test = X_test/255
```

```python
#Noise Addition
noise_factor = 0.3
noise_dataset, noise_test_dataset = [], []

for img in X_train:
    noisy_image = img + noise_factor*np.random.randn(*img.shape)
    noisy_image = np.clip(noisy_image, 0, 1)
    noise_dataset.append(noisy_image)
    

for img in X_test:
    noisy_image = img + noise_factor*np.random.randn(*img.shape)
    noisy_image = np.clip(noisy_image, 0, 1)
    noise_test_dataset.append(noisy_image)
    
noise_dataset = np.array(noise_dataset)
noise_test_dataset = np.array(noise_test_dataset)
```

## Step 4 : Build and Train Model

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, Conv2DTranspose
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from livelossplot import PlotLossesKerasTF
```

```python
ae = Sequential()

#Encoder
ae.add(Conv2D(filters=16, kernel_size=3, strides=2, padding="same", input_shape=(28,28,1)))
ae.add(Conv2D(filters=8, kernel_size=3, strides=2, padding="same"))

ae.add(Conv2D(filters=8, kernel_size=3, strides=1, padding="same"))

#Decoder
ae.add(Conv2DTranspose(filters=16, kernel_size=3, strides=2, padding="same"))
ae.add(Conv2DTranspose(filters=1, kernel_size=3, strides=2, padding="same", activation="sigmoid"))

ae.compile(loss="binary_crossentropy", optimizer="adam")
ae.summary()
```

```python
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)
early_stopping = EarlyStopping(monitor='val_loss', patience=5)
callbacks = [PlotLossesKerasTF(), checkpoint, early_stopping]

ae_hist = ae.fit(noise_dataset.reshape(-1,28,28,1), X_train.reshape(-1,28,28,1), epochs=100, batch_size=128, validation_data=(noise_test_dataset.reshape(-1,28,28,1), X_test.reshape(-1,28,28,1)), callbacks=callbacks)
```

## Step 5 : Evaluate Model Performance

```python
ae = load_model("best_model.h5")
```

```python
accuracy = ae.evaluate(noise_test_dataset.reshape(-1,28,28,1), X_test.reshape(-1,28,28,1))
print("Test Accuracy : {:.3f}".format(accuracy))
```
